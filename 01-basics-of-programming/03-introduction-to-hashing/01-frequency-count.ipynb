# Frequency of an Element

## Problem Statement

Given an array `arr` of positive integers and an integer `x`. Return the frequency of `x` in the array.

---

## Examples

### Example 1

**Input:**  
`arr = [1, 1, 1, 1, 1]`, `x = 1`

**Output:**  
`5`

**Explanation:** Frequency of 1 is 5.

---

### Example 2

**Input:**  
`arr = [1, 2, 3, 3, 2, 1]`, `x = 2`

**Output:**  
`2`

**Explanation:** Frequency of 2 is 2.

---

## Constraints

| Constraint | Value |
|------------|-------|
| 1 ≤ arr.size() ≤ 10^5 | Number of elements |
| 1 ≤ arr[i] ≤ 10^5 | Array element values |
| 1 ≤ x ≤ 10^5 | Target element |

```python
def findFrequency(arr : list, x: int) -> int:
    count = 0
    for e in arr:
        if e == x:
            count += 1
    
    return count   

l = [1,4,5,5,2,5,6,9]
result = findFrequency(l, 5)
print(result)
```

**Time Complexity** = `O(n)`

**Space Complexity** = `O(1)`
