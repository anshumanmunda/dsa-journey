# Armstrong Number

## Problem Statement

Given an integer `n`, write a program to check whether the number is an **Armstrong number** or not.

A number is called an Armstrong number if the sum of its digits each raised to the power of the number of digits equals the number itself.

---

## Examples

### Example 1

**Input:**  
`n = 153`

**Output:**  
`true` (or `"Armstrong"`)

**Explanation:**  
`1³ + 5³ + 3³ = 1 + 125 + 27 = 153` → Yes, it is an Armstrong number.

---

### Example 2

**Input:**  
`n = 9474`

**Output:**  
`true` (or `"Armstrong"`)

**Explanation:**  
`9⁴ + 4⁴ + 7⁴ + 4⁴ = 6561 + 256 + 2401 + 256 = 9474` → Yes, it is an Armstrong number.

---

### Example 3

**Input:**  
`n = 123`

**Output:**  
`false` (or `"Not Armstrong"`)

**Explanation:**  
`1³ + 2³ + 3³ = 1 + 8 + 27 = 36` → 36 ≠ 123, so it is not an Armstrong number.

---

## Constraints

| Constraint | Value |
|------------|-------|
| 1 <= n <= 10^6 | n is a positive integer |

---

## Problem Summary

| Aspect | Details |
|--------|---------|
| **Task** | Check if the given number is an Armstrong number |
| **Input** | Integer `n` |
| **Output** | `true` / `false` or `"Armstrong"` / `"Not Armstrong"` |

```python
def isArmstrong(num: int) -> bool:
  if num < 0:
    return
  temp = num
  no_of_digit = 0

  while num:      # number of digits 
    num //=  10
    no_of_digit += 1

  num = temp
  sum = 0 
  while num:           # sum of  digit each rase to power no of digits
    digit = num % 10
    sum +=  digit ** no_of_digit
    num //= 10

  if sum == temp: 
    print(f'{temp} is an Armstrong number') 
  else:  
    print(f'{temp} is not an  Armstrong number') 

isArmstrong(123)
```

**Time Complexity** = `O(n)+O(n) = O(n)`

**Space Complexity** = `O(1)`
