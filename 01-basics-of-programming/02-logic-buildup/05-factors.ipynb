# Number of Factors

## Problem Statement

Find the number of factors for a given integer `n`.

---

## Examples

### Example 1

**Input:**  
`n = 5`

**Output:**  
`2`

**Explanation:** 5 has 2 factors: 1 and 5.

---

### Example 2

**Input:**  
`n = 25`

**Output:**  
`3`

**Explanation:** 25 has 3 factors: 1, 5, and 25.

---

## Constraints

| Constraint | Value |
|------------|-------|
| 1 ≤ n ≤ 10^5 | n is a positive integer |

```python
# Brute Force Approach

def countFactors (n:int) -> int:
  factors = 0
  for e in range (1, n+1):
      if n % e == 0:
          factors += 1
  return factors        

result = countFactors(20)
print(result)
```

**Time Complexity** = `O(n)`

**Space Complexity** = `O(1)`

```python
# Better Approach

def countFactors (n:int) -> int:
  factors = 2
  k = n // 2
  for e in range (2, k+1):
      if n % e == 0:
          factors += 1
  return factors        

result = countFactors(20)
print(result)
```

**Time Complexity** = `O(n/2) => O(n)`

**Space Complexity** = `O(1)`
