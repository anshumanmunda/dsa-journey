# Number of Digits

## Problem Statement

Ninja wants to add coding to his skill set, so he started learning it. On the first day, he got stuck on a problem where he was given a long integer **X** and had to count the number of digits in it.

Ninja called you for help as you are his only friend. Help him solve the problem.

---

## Example

**Input:**  
`X = 2`

**Output:**  
`1`

> Only one digit `2` is present in `X`, so the answer is `1`.

---

## Constraints

| Constraint | Value |
|------------|-------|
| T (Test cases) | 1 <= T <= 1000 |
| X (Integer) | 1 <= X <= 10^18 |
| Time Limit | 1 second |

---

## Sample Input / Output

### Sample Input 1


### Explanation of Sample Input 1

| Test Case | Input | Digits Present | Output |
|-----------|-------|----------------|--------|
| 1 | 89 | 8, 9 | 2 |
| 2 | 870 | 8, 7, 0 | 3 |



```python
def countDigit(n: int) -> int:
   count = 0
   while n:  
      n //= 10
      count += 1 
   return count  

result = countDigit(125126)
print(f'Number of digits = {result}')
```

>**Time complexity** = `O(log n)` <br>
>**Space complexity** = `O(1)`
